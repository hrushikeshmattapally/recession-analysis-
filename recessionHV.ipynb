```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')

import hvplot.pandas
```

```python
df1=pd.read_csv("NSEI.csv")
df1
```

```python
line_chart1=df1.hvplot.line(x='Date', y='High')
line_chart1
```

```python
nifty50_df = df1.loc[[0,10,30,40,50,60,70,80,90,100,110,120,130,140,150,160,167]]
print(nifty50_df)
```

```python

line_chart2 = nifty50_df.hvplot(x='Date', 
                         y=['High'],
                         title='NIFTY 50 Share Price Chart ',
                         xlabel='Date',
                         ylabel='share price',
                         line_width=2.5,
                         legend='bottom_left',
                         grid=True,
                         width=800,
                         height=400,rot=60)
line_chart2
```

```python
df2=pd.read_csv("global_inflation_data.csv")
df2
```

```python
Indian_inflation=df2[df2["country_name"]=="India"]
Indian_inflation
```

```python
years = [col for col in Indian_inflation.keys() if col.isdigit()]
df = pd.DataFrame(Indian_inflation) # Create the DataFrame
df = df.melt(id_vars=["country_name", "indicator_name"], var_name="years", value_name="value")#Select the desired columns and reshape
inflation_data=df
print(inflation_data)
```

```python
bar_chart1=inflation_data.hvplot.bar(x='years', y='value', rot=90)
bar_chart1 = bar_chart1.opts(title=" Indian Inflation Rate Over the Years")
bar_chart1
```

```python
df=pd.read_csv("indian_GDP_growth.csv")
df3= df[df['Year'] > 1960]
df3
```

```python
bar_chart2=df3.hvplot.bar(x='Year', y='Percentage_Growth ', rot=90)
bar_chart2
```

```python
df=pd.read_csv('indian_GDP_growth_updated2.csv')
df4= df[df['Year'] > 1989]
df4
```

```python
bar_chart3=df4.hvplot.bar(x='Year', y='Percentage_Growth_usa', rot=90)
bar_chart3 = bar_chart3.opts(title="G.D.P growth rate of U.S.A")

bar_chart3
```

```python
bar_chart6=df4.hvplot.bar(x='Year', y='Percentage_Growth_india', rot=90)
bar_chart6 = bar_chart6.opts(title="G.D.P growth rate of india")
bar_chart6
```

```python
line_chart3=df4.hvplot(x='Year', y=['Percentage_Growth_india', 'Percentage_Growth_usa','Percentage_Growth_japan'],
             value_label='Rate (per 100k people)')
line_chart3
```

```python

line_chart4 = df4.hvplot(
    x='Year',
    y=['Percentage_Growth_india', 'Percentage_Growth_usa', 'Percentage_Growth_japan'],
    title="Economic Growth Comparison: India, USA, Japan (20XX-20XX)",
    xlabel='Year',
    legend=True,  # Add a legend
)
line_chart4
```

```python
line_chart5 = df4.hvplot(x='Year', 
                         y=['Percentage_Growth_india', 'Percentage_Growth_usa', 'Percentage_Growth_japan'],
                         value_label='Rate (per 100k people)',
                         title='G.D.P Growth Comparison',
                         xlabel='Year',
                         ylabel='Percentage Growth',
                         line_color=['red', 'blue', 'green'],
                         line_width=2.5,
                         legend='bottom_left',
                         grid=True,
                         width=800,
                         height=400)
line_chart5
```

```python
import holoviews as hv
line_chart7 = df4.hvplot(
    x='Year',
    y=['Percentage_Growth_india', 'Percentage_Growth_usa',
       'Percentage_Growth_japan', 'Percentage_growth_china', 'Percentage_Growth_uk'],
    value_label='Rate (per 100k people)',
    title='G.D.P Growth Comparison',
    xlabel='Year',
    ylabel='Percentage Growth',
    line_color=['blue', 'orange', 'gray', 'red', 'black'],
    line_width=2.5,
    # Set legend position to 'right' for horizontal placement
    legend='right',  # Maintain this for horizontal positioning
    grid=True,
    width=800,
    height=400
)

# Adjust legend position further using Bokeh-supported options:
line_chart7 = line_chart7.opts(
    # Move the legend slightly to the left to avoid overlapping lines
    legend_offset=(10, 0),  # Adjust as needed
    # Control the spacing between legend entries (optional)
    legend_spacing=10   # Adjust as needed
)

line_chart7
```
